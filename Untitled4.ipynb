```python
class Persona:
    def __init__(self, nombre="", edad=0, direccion=""):
        self.nombre = nombre
        self.edad = edad
        self.direccion = direccion

    def imprime(self):
        print("Nombre:", self.nombre)
        print("Edad:", self.edad)
        print("Dirección:", self.direccion)


class Cliente(Persona):
    def __init__(self, nombre="", edad=0, direccion="", numCliente=0, cel=0, servicio=""):
        super().__init__(nombre, edad, direccion)
        self.numCliente = numCliente
        self.cel = cel
        self.servicio = servicio

    def imprime(self):
        super().imprime()
        print("NumCliente:", self.numCliente)
        print("Cel:", self.cel)
        print("Servicio:", self.servicio)


# --- Parte 3: pedir datos ---
nombre = input("Nombre: ")
edad = int(input("Edad: "))
direccion = input("Dirección: ")
numCliente = input("Número de cliente: ")
cel = int(input("Celular: "))
servicio = input("Servicio: ")

cliente = Cliente(nombre, edad, direccion, numCliente, cel, servicio)

print("\n--- Datos del cliente ---")
cliente.imprime()
```
